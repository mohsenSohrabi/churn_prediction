#### Functions to demonstrate categorical data
We need a function that shows value counts for categorical data. 

```python
def show_value_counts(df, column, figsize=(10, 6)):
    # Calculate the value counts for the specified column
    value_counts = df[column].value_counts()
    
    # Display the value counts
    print(value_counts)
    
    # Create a bar chart showing the distribution of values
    plt.figure(figsize=figsize)
    sns.countplot(data=df, x=column, order=value_counts.index)
    plt.show()
```

## Target feature 

**churn**: The `churn` (the last column in the dataset)column is the target variable for a churn prediction problem. It indicates whether a customer has churned (left the company) or not. The values in this column are either ‘**yes**’ or ‘**no**’, where ‘**yes**’ indicates that the customer has churned and ‘**no**’ indicates that the customer has not churned. 

#### Value distribution for churn 

```python
show_value_counts(data, 'churn')
```

The result  provided shows that the `churn` column in the dataset is imbalanced, with many more instances of customers who did not `churn` (no) than customers who did churn (yes). This can pose a challenge when building a predictive model, as the model may simply predict the majority class (no) for all instances, resulting in poor performance.

There are several techniques that can be used to address class imbalance in a dataset. Some common approaches include:
* **DownSampling (Undersampling)**: This involves randomly removing instances of the majority class from the dataset to balance the class distribution. This can help to prevent the model from being biased towards the majority class, but it also means that potentially useful data is discarded.
  
* **Oversampling**: This involves randomly duplicating instances of the minority class in the dataset to balance the class distribution. This can help to prevent the model from being biased towards the majority class, but it can also lead to overfitting, as the model may simply memorize the duplicated instances.

* **Synthetic data generation**: This involves generating new synthetic instances of the minority class using techniques such as SMOTE (Synthetic Minority Over-sampling Technique). This can help to balance the class distribution without discarding any data or introducing overfitting, but it requires careful implementation to ensure that the synthetic data is representative of the underlying distribution.

**We will use upsampling later to fix the imbalanced data problem**. 


### Non-target columns(features)
We can divide our features into two groups: categorical and non-categorical fetaures. Categorical data takes data in a limited range while non-categorical data often take numbers in wide range. 

#### Categorical Columns
To determine the significance of categorical features, we should examine their different values and display their distribution or proportion in relation to the target features. This will help us understand the importance of each categorical feature. For this purpose we will use the below function.

```python
def plot_categorical_proportions(data, cat_col, target_col, figsize=(15, 6)):
    # Group the data by the categorical and target columns and calculate the size of each group
    grouped_data = data.groupby([cat_col, target_col]).size().reset_index(name='count')

    # Calculate the total number of instances for each category
    cat_totals = grouped_data.groupby(cat_col)['count'].sum().reset_index(name='total')

    # Merge the grouped data with the category totals
    grouped_data = pd.merge(grouped_data, cat_totals, on=cat_col)

    # Calculate the proportion of instances in each target class for each category
    grouped_data['proportion'] = grouped_data['count'] / grouped_data['total']

    # Create a bar chart showing the proportion of instances in each target class for each category
    plt.figure(figsize=figsize)
    sns.barplot(data=grouped_data, x=cat_col, y='proportion', hue=target_col)
    plt.xticks(rotation=90)
    plt.show()
```

This function takes a pandas DataFrame `data`, the names of the categorical and target columns (`cat_col` and `target_col`), and an optional figure size figsize as input. It performs a grouped analysis to calculate the proportion of instances in each target class for each category in the specified categorical column. The resulting proportions are then displayed in a bar chart using the `barplot()` function from the `seaborn` library. The x-axis labels are rotated by 90 degrees to make them more readable.

### Non-Catregorical features
Working with non-categorical data we often start by creating visualizations such as histograms or box plots to understand the distribution of the data. This can help you identify any outliers or unusual patterns in the data. Here is the function we can use to create box plot for numerical data. 

```python
def describe_and_plot_numerical(data, column):
    # Calculate summary statistics for the specified column
    summary = data[column].describe()
    
    # Display the summary statistics
    print(summary)
    
    # Create a box plot to visualize the distribution of the data
    plt.figure()
    sns.boxplot(data=data, x=column)
    plt.show()
```

Similar to what we have done for categorical data we can visualize the distribution of numerical data according to the target feature by creating separate box plots for each value in our target field. In our case, our target feature is `churn`, which has two values `yes` and `no`, we can create two box plots to show the distribution of a numerical feature for customers who churned (yes) and customers who did not churn (no). This can help us compare the distributions and identify any differences between the two groups. Here is the function:

```python
def plot_numerical_by_target(data, numerical_col, target_col='churn'):
    # Create a box plot to visualize the distribution of the numerical column for each value of the target column
    plt.figure()
    sns.boxplot(data=data, x=target_col, y=numerical_col)
    plt.show()
```

We can use a density function to compare the distribution of data. However, if the data is imbalanced, we may see two plots with one being larger than the other. So it is less useful than the box plot. 

```python
def plot_numerical_density_by_target(data, numerical_col, target_col):
    # Create a density plot to visualize the distribution of the numerical column for each value of the target column
    plt.figure()
    sns.kdeplot(data=data, x=numerical_col, hue=target_col)
    plt.show()
```
